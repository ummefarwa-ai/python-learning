```python
# ==========================================
# Seaborn - Histogram (Histplot)
# ==========================================

import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
tips = sns.load_dataset("tips")
```

```python
# ==========================================
# Basic Histogram
# ==========================================
plt.figure(figsize=(7,5))

sns.histplot(
    data=tips,
    x="total_bill"
)

plt.title("Distribution of Total Bill")

plt.show()
```

```python
# ==========================================
# Change Number of Bins
# ==========================================
plt.figure(figsize=(7,5))

sns.histplot(
    data=tips,
    x="total_bill",
    bins=20
)

plt.title("Histogram with 20 Bins")

plt.show()
```

```python
# ==========================================
# Add KDE Curve
# ==========================================

plt.figure(figsize=(7,5))

sns.histplot(
    data=tips,
    x="total_bill",
    kde=True
)

plt.title("Histogram with KDE")

plt.show()
```

```python
# ==========================================
# Change Color
# ==========================================
plt.figure(figsize=(7,5))

sns.histplot(
    data=tips,
    x="tip",
    color="green",
    kde=True
)

plt.title("Tip Distribution")

plt.show()
```

```python
# ==========================================
# Histogram with hue
# ==========================================

plt.figure(figsize=(8,5))

sns.histplot(
    data=tips,
    x="total_bill",
    hue="sex",
    kde=True
)

plt.title("Bill Distribution by Gender")

plt.show()
```

```python
# ==========================================
# Multiple Histograms
# ==========================================

plt.figure(figsize=(8,5))

sns.histplot(
    data=tips,
    x="total_bill",
    hue="time",
    multiple="layer"
)

plt.title("Lunch vs Dinner Bills")

plt.show()
```

```python
# ==========================================
# SFilled Histogram
# ==========================================
plt.figure(figsize=(8,5))

sns.histplot(
    data=tips,
    x="tip",
    hue="smoker",
    multiple="fill"
)

plt.title("Tip Distribution by Smoking Status")

plt.show()
```

```python
# ==========================================
# Professional Example
# ==========================================

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8,5))

sns.histplot(
    data=tips,
    x="total_bill",
    kde=True,
    color="steelblue",
    bins=25
)

plt.title("Distribution of Restaurant Bills")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")

plt.show()
```
