```python
# ==========================================
# Seaborn - Violin Plot
# ==========================================

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Load Dataset
tips = sns.load_dataset("tips")
```

```python
# ==========================================
# Basic Violin Plot
# ==========================================

plt.figure(figsize=(7,5))

sns.violinplot(
    data=tips,
    y="total_bill"
)

plt.title("Distribution of Total Bill")

plt.show()
```

```python
# ==========================================
# Horizontal Violin Plot
# ==========================================
plt.figure(figsize=(7,5))

sns.violinplot(
    data=tips,
    x="tip"
)

plt.title("Distribution of Tips")

plt.show()
```

```python
# ==========================================
# Violin Plot by Category
# ==========================================

plt.figure(figsize=(8,5))

sns.violinplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.title("Total Bill by Day")

plt.show()
```

```python
# ==========================================
# Violin Plot with hue
# ==========================================

plt.figure(figsize=(8,5))

sns.violinplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)

plt.title("Total Bill by Day and Gender")

plt.show()
```

```python
# ==========================================
# Split Violin Plot
# ==========================================

plt.figure(figsize=(8,5))

sns.violinplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    split=True
)

plt.title("Split Violin Plot")

plt.show()
```

```python
# ==========================================
# Inner Quartiles
# ==========================================
plt.figure(figsize=(8,5))

sns.violinplot(
    data=tips,
    x="time",
    y="tip",
    inner="quart"
)

plt.title("Tips by Meal Time")

plt.show()
```

```python
# ==========================================
# Compare Smokers and Non-Smokers
# ==========================================

plt.figure(figsize=(8,5))

sns.violinplot(
    data=tips,
    x="smoker",
    y="tip",
    palette="Set2"
)

plt.title("Tips by Smoking Status")

plt.show()
```

```python
# ==========================================
# Professional Example
# ==========================================
sns.set_theme(style="whitegrid")

plt.figure(figsize=(9,5))

sns.violinplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="time",
    palette="viridis"
)

plt.title("Restaurant Bill Distribution")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")

plt.show()
```
