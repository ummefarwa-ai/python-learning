```python
# ==========================================
# Seaborn - Count Plot
# ==========================================

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')


# Load Dataset
tips = sns.load_dataset("tips")
```

```python
# ==========================================
# Basic Count Plot
# ==========================================
plt.figure(figsize=(7,5))

sns.countplot(
    data=tips,
    x="day"
)

plt.title("Number of Customers by Day")

plt.show()
```

```python
# ==========================================
# Count Plot with Color
# ==========================================

plt.figure(figsize=(7,5))

sns.countplot(
    data=tips,
    x="sex",
    color="green"
)

plt.title("Gender Distribution")

plt.show()
```

```python
# ==========================================
# Count Plot with Palette
# ==========================================

plt.figure(figsize=(7,5))

sns.countplot(
    data=tips,
    x="day",
    palette="Set2"
)

plt.title("Customers by Day")

plt.show()
```

```python
# ==========================================
# Count Plot with hue
# ==========================================

plt.figure(figsize=(8,5))

sns.countplot(
    data=tips,
    x="day",
    hue="sex"
)

plt.title("Customers by Day and Gender")

plt.show()
```

```python
# ==========================================
# Horizontal Count Plot
# ==========================================

plt.figure(figsize=(8,5))

sns.countplot(
    data=tips,
    y="day",
    palette="pastel"
)

plt.title("Horizontal Count Plot")

plt.show()
```

```python
# ==========================================
# Count Smokers and Non-Smokers
# ==========================================
plt.figure(figsize=(7,5))

sns.countplot(
    data=tips,
    x="smoker",
    palette="viridis"
)

plt.title("Smokers vs Non-Smokers")

plt.show()
```

```python
# ==========================================
# Count Plot with Multiple Categories
# ==========================================
plt.figure(figsize=(8,5))

sns.countplot(
    data=tips,
    x="time",
    hue="smoker",
    palette="coolwarm"
)

plt.title("Meal Time by Smoking Status")

plt.show()
```

```python
# ==========================================
# Professional Example
# ==========================================

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8,5))

sns.countplot(
    data=tips,
    x="day",
    hue="time",
    palette="Set1"
)

plt.title("Customer Visits by Day and Meal Time")
plt.xlabel("Day")
plt.ylabel("Number of Customers")

plt.show()
```
