```python
# ==========================================
# Seaborn - Box Plot
# ==========================================

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Load Dataset
tips = sns.load_dataset("tips")
```

```python
# ==========================================
# Basic Box Plot
# ==========================================

plt.figure(figsize=(7,5))

sns.boxplot(
    data=tips,
    y="total_bill"
)

plt.title("Distribution of Total Bill")

plt.show()
```

```python
# ==========================================
# Horizontal Box Plot
# ==========================================
plt.figure(figsize=(7,5))

sns.boxplot(
    data=tips,
    x="tip"
)

plt.title("Distribution of Tips")

plt.show()
```

```python
# ==========================================
# Box Plot by Category
# ==========================================

plt.figure(figsize=(8,5))

sns.boxplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.title("Total Bill by Day")

plt.show()
```

```python
# ==========================================
# Box Plot with hue
# ==========================================

plt.figure(figsize=(8,5))

sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)

plt.title("Total Bill by Day and Gender")

plt.show()
```

```python
# ==========================================
# Change Color Palette
# ==========================================

plt.figure(figsize=(8,5))

sns.boxplot(
    data=tips,
    x="time",
    y="tip",
    palette="Set2"
)

plt.title("Tips by Meal Time")

plt.show()
```

```python
# ==========================================
# Show Outliers Clearly
# ==========================================
plt.figure(figsize=(8,5))

sns.boxplot(
    data=tips,
    y="total_bill",
    color="lightgreen"
)

plt.title("Detecting Outliers")

plt.show()
```

```python
# ==========================================
# Compare Smokers and Non-Smokers
# ==========================================

plt.figure(figsize=(8,5))

sns.boxplot(
    data=tips,
    x="smoker",
    y="tip",
    palette="pastel"
)

plt.title("Tips by Smoking Status")

plt.show()
```

```python
# ==========================================
# Professional Example
# ==========================================

sns.set_theme(style="whitegrid")

plt.figure(figsize=(9,5))

sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="time",
    palette="viridis"
)

plt.title("Restaurant Bill Distribution")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")

plt.show()
```
