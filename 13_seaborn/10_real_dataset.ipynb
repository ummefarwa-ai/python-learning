```python
# ==========================================
# Seaborn - Real Dataset Visualization
# ==========================================

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Set Theme
sns.set_theme(style="whitegrid")

# Load Dataset
tips = sns.load_dataset("tips")
```

```python
# ==========================================
# Display the Dataset
# ==========================================


tips.head()
```

```python
# ==========================================
# Dataset Information
# ==========================================


tips.info()
```

```python
# ==========================================
# Statistical Summary
# ==========================================


tips.describe()
```

```python
# ==========================================
# Bar Plot
# ==========================================


plt.figure(figsize=(8,5))

sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    palette="viridis"
)

plt.title("Average Total Bill by Day")

plt.show()
```

```python
# ==========================================
# Scatter Plot
# ==========================================


plt.figure(figsize=(8,5))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)

plt.title("Total Bill vs Tip")

plt.show()
```

```python
# ==========================================
# Box Plot
# ==========================================


plt.figure(figsize=(8,5))

sns.boxplot(
    data=tips,
    x="day",
    y="tip",
    palette="Set2"
)

plt.title("Tips by Day")

plt.show()
```

```python
# ==========================================
# Histogram
# ==========================================


plt.figure(figsize=(8,5))

sns.histplot(
    data=tips,
    x="total_bill",
    kde=True,
    color="steelblue"
)

plt.title("Distribution of Total Bill")

plt.show()
```

```python
# ==========================================
# Correlation Heatmap
# ==========================================


numeric_data = tips.select_dtypes(include="number")

plt.figure(figsize=(7,5))

sns.heatmap(
    numeric_data.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```

```python
# ==========================================
# Pair Plot
# ==========================================


sns.pairplot(
    tips,
    vars=["total_bill", "tip", "size"],
    hue="time",
    palette="Set2"
)

plt.show()
```
