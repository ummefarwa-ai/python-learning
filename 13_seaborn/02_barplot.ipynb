```python
# ==========================================
# Seaborn - Bar Plot
# ==========================================

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset("tips")
```

```python
# ==========================================
# Basic Bar Plot
# ==========================================
plt.figure(figsize=(7,5))

sns.barplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.title("Average Total Bill by Day")

plt.show()
```

```python
# ==========================================
# Change Color
# ==========================================
plt.figure(figsize=(7,5))

sns.barplot(
    data=tips,
    x="day",
    y="tip",
    color="skyblue"
)

plt.title("Average Tip by Day")

plt.show()
```

```python
# ==========================================
# Group Data Using hue
# ==========================================

plt.figure(figsize=(7,5))

sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)

plt.title("Average Bill by Gender")

plt.show()
```

```python
# ==========================================
# Horizontal Bar Plot
# ==========================================

plt.figure(figsize=(7,5))

sns.barplot(
    data=tips,
    x="tip",
    y="day",
    color="lightgreen"
)

plt.title("Average Tip by Day")

plt.show()
```

```python
# ==========================================
# Change Color Palette
# ==========================================

plt.figure(figsize=(7,5))

sns.barplot(
    data=tips,
    x="day",
    y="tip",
    palette="Set2"
)

plt.title("Tips using Set2 Palette")

plt.show()
```

```python
# ==========================================
# Display Error Bars
# ==========================================

plt.figure(figsize=(7,5))

sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    errorbar="sd"
)

plt.title("Average Bill with Standard Deviation")

plt.show()
```

```python
# ==========================================
# Compare Smokers and Non-Smokers
# ==========================================

plt.figure(figsize=(8,5))

sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="smoker",
    palette="pastel"
)

plt.title("Average Tip by Smoking Status")

plt.show()
```

```python
# ==========================================
# Professional Example
# ==========================================
sns.set_theme(style="whitegrid")

plt.figure(figsize=(8,5))

sns.barplot(
    data=tips,
    x="time",
    y="total_bill",
    hue="sex",
    palette="viridis"
)

plt.title("Average Total Bill by Meal Time and Gender")
plt.xlabel("Meal Time")
plt.ylabel("Average Total Bill")

plt.show()
```
