```python
# ==========================================
# Seaborn - Pair Plot
# ==========================================

import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
tips = sns.load_dataset("tips")
```

```python
# ==========================================
# Basic Pair Plot
# ==========================================

sns.pairplot(tips)

plt.show()
```

```python
# ==========================================
# Pair Plot with hue
# ==========================================

sns.pairplot(
    tips,
    hue="sex"
)

plt.show()
```

```python
# ==========================================
# Different Marker Styles
# ==========================================

sns.pairplot(
    tips,
    hue="smoker",
    markers=["o", "s"]
)

plt.show()
```

```python
# ==========================================
# Histogram on the Diagonal
# ==========================================

sns.pairplot(
    tips,
    diag_kind="hist"
)

plt.show()
```

```python
# ==========================================
# KDE on the Diagonal
# ==========================================

sns.pairplot(
    tips,
    diag_kind="kde"
)

plt.show()
```

```python
# ==========================================
# Select Specific Columns
# ==========================================

sns.pairplot(
    tips,
    vars=["total_bill", "tip", "size"]
)

plt.show()
```

```python
# ==========================================
# Pair Plot with Color Palette
# ==========================================

sns.pairplot(
    tips,
    hue="day",
    palette="Set2"
)

plt.show()
```

```python
# ==========================================
# Professional Example
# ==========================================

sns.set_theme(style="ticks")

sns.pairplot(
    tips,
    vars=["total_bill", "tip", "size"],
    hue="time",
    diag_kind="hist",
    palette="viridis"
)

plt.show()
```
