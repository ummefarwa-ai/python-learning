```python
# ==========================================
# Seaborn - Heatmap
# ==========================================

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Load Dataset
tips = sns.load_dataset("tips")
```

```python
# ==========================================
# Correlation Matrix
# ==========================================

# Select only numeric columns
numeric_data = tips.select_dtypes(include="number")

# Calculate correlation
correlation = numeric_data.corr()

print(correlation)
```

```python
# ==========================================
# Basic Heatmap
# ==========================================

plt.figure(figsize=(7,5))

sns.heatmap(correlation)

plt.title("Correlation Heatmap")

plt.show()
```

```python
# ==========================================
# Display Correlation Value
# ==========================================

plt.figure(figsize=(7,5))

sns.heatmap(
    correlation,
    annot=True
)

plt.title("Correlation Heatmap with Values")

plt.show()
```

```python
# ==========================================
# Change Color Map
# ==========================================

plt.figure(figsize=(7,5))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm"
)

plt.title("Coolwarm Heatmap")

plt.show()
```

```python
# ==========================================
# Add Grid Lines
# ==========================================

plt.figure(figsize=(7,5))

sns.heatmap(
    correlation,
    annot=True,
    cmap="YlGnBu",
    linewidths=1,
    linecolor="white"
)

plt.title("Heatmap with Grid Lines")

plt.show()
```

```python
# ==========================================
# Display Only Lower Triangle
# ==========================================

import numpy as np

mask = np.triu(np.ones_like(correlation, dtype=bool))

plt.figure(figsize=(7,5))

sns.heatmap(
    correlation,
    mask=mask,
    annot=True,
    cmap="viridis"
)

plt.title("Lower Triangle Correlation")

plt.show()
```

```python
# ==========================================
# Professional Heatmap
# ==========================================

sns.set_theme(style="white")

plt.figure(figsize=(8,6))

sns.heatmap(
    correlation,
    annot=True,
    cmap="RdYlBu",
    linewidths=0.5,
    square=True
)

plt.title("Correlation Matrix")

plt.show()
```

```python
# ==========================================
# Heatmap with Custom Value Format
# ==========================================

plt.figure(figsize=(8,6))

sns.heatmap(
    correlation,
    annot=True,
    fmt=".2f",
    cmap="Blues"
)

plt.title("Correlation Values (2 Decimal Places)")

plt.show()
```

```python
# ==========================================
# Heatmap for a Confusion Matrix
# ==========================================

import numpy as np

conf_matrix = np.array([
    [45, 5],
    [3, 47]
])

plt.figure(figsize=(5,4))

sns.heatmap(
    conf_matrix,
    annot=True,
    fmt="d",
    cmap="Greens",
    xticklabels=["Predicted No", "Predicted Yes"],
    yticklabels=["Actual No", "Actual Yes"]
)

plt.title("Confusion Matrix")

plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")

plt.show()
```
