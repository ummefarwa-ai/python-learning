```python
# ==========================================
# Seaborn - Introduction
# ==========================================

# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# ==========================================
# Check Seaborn Version
# ==========================================
print("Seaborn Version:", sns.__version__)
```

```python
# ==========================================
# Load a Built-in Dataset
# ==========================================
tips = sns.load_dataset("tips")

tips.head()
```

```python
# ==========================================
# Dataset Information
# ==========================================
tips.info()
```

```python
# ==========================================
# Statistical Summary
# ==========================================

tips.describe()
```

```python
# ==========================================
# Display Dataset Shape
# ==========================================
print("Rows and Columns:", tips.shape)
```

```python
# ==========================================
# Display Column Names
# ==========================================

print(tips.columns)
```

```python
# ==========================================
# Set Seaborn Theme
# ==========================================

sns.set_theme(style="darkgrid")

plt.figure(figsize=(7,5))

sns.lineplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.title("Total Bill vs Tip")

plt.show()
```

```python
# ==========================================
# Try Different Themes
# ==========================================
sns.set_theme(style="whitegrid")

plt.figure(figsize=(7,5))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.title("Whitegrid Theme")

plt.show()
```

```python
# ==========================================
# Change Color Palette
# ==========================================
sns.set_palette("Set2")

plt.figure(figsize=(7,5))

sns.barplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.title("Average Bill by Day")

plt.show()
```

```python
# ==========================================
# Professional Example
# ==========================================

sns.set_theme(style="ticks")

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker",
    s=100
)

plt.title("Restaurant Tips Dataset")

plt.show()
```
