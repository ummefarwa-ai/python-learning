```python
# ==========================================
# Seaborn - Scatter Plot
# ==========================================

import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
tips = sns.load_dataset("tips")
```

```python
# ==========================================
# Basic Scatter Plot
# ==========================================

plt.figure(figsize=(7,5))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.title("Total Bill vs Tip")

plt.show()
```

```python
# ==========================================
# Change Marker Color
# ==========================================

plt.figure(figsize=(7,5))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    color="red"
)

plt.title("Scatter Plot with Red Markers")

plt.show()
```

```python
# ==========================================
# Color Points Using hue
# ==========================================
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex"
)

plt.title("Total Bill vs Tip by Gender")

plt.show()
```

```python
# ==========================================
# Different Marker Styles
# ==========================================

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    style="smoker"
)

plt.title("Scatter Plot by Smoking Status")

plt.show()
```

```python
# ==========================================
# Change Marker Size
# ==========================================

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    size="size"
)

plt.title("Marker Size Represents Group Size")

plt.show()
```

```python
# ==========================================
# Combine hue, style, and size
# ==========================================

plt.figure(figsize=(9,6))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    style="smoker",
    size="size"
)

plt.title("Multiple Variables in One Scatter Plot")

plt.show()
```

```python
# ==========================================
# Change Color Palette
# ==========================================

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time",
    palette="Set2"
)

plt.title("Lunch vs Dinner Bills")

plt.show()
```

```python
# ==========================================
# Professional Example
# ==========================================

sns.set_theme(style="whitegrid")

plt.figure(figsize=(9,6))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    style="sex",
    size="size",
    palette="viridis",
    sizes=(40, 200)
)

plt.title("Restaurant Tips Analysis")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

plt.show()
```
