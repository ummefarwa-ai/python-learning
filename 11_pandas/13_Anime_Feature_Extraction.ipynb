**make a new column for episode count**

```python
# make a new column for episode count   but we dont have episode count feature ===>so we will do extracting feature
df.loc[2]["Title"]
```

```python
# create a function to extract episode counts
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt: 
        if i == ")":
             check = False
             return data
    
        if check == True:
            data = data + i 
        if i =="(":
            check = True
            
   
```

```python
df["Episodes"] = df["Title"].apply(extract_episodes)
```

```python
df["Episodes"] = df["Episodes"].str.replace(" eps", "")
df["Episodes"] = df["Episodes"].astype(int)
```

```python
# creating a new function for extraction time stamp feature from title 
def time_stamp(txt):
    check = False
    data = ""
    for i in txt:
        if i == ",":
            check = False 
            return data
        if check == True:
            data = data + i
        if i == ")":
            check = True
```

```python
# better way to do it 
def extract_date(txt):
    check = False
    data= "" 
    for i in range(len(txt)): 
        if txt[i] == ")":
            for j in range(i+1, i+20):
                data += txt[j]
            return data
```

```python
df["Duration"] = df["Title"].apply(extract_date)
```

```python
df.head()
```

```python
# Convert duration into start and end dates
dates = df["Duration"].str.split(" - ", expand=True)

start = pd.to_datetime(dates[0], format="%b %Y")
end = pd.to_datetime(dates[1], format="%b %Y")

# Calculate difference in months
df["Months"] = (
    (end.dt.year - start.dt.year) * 12
    + (end.dt.month - start.dt.month)
    + 1
)
df.head()
```
