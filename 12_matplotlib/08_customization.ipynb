```python
# ==========================================
# Matplotlib - Plot Customization
# ==========================================

import matplotlib.pyplot as plt
import numpy as np
```

```python
# ==========================================
# Customize Title and Axis Labels
# ==========================================
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 18, 25]

plt.plot(x, y)

plt.title("Monthly Sales", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Sales", fontsize=12)

plt.show()
```

```python
# ==========================================
# Change Figure Size
# ==========================================
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.figure(figsize=(10,5))

plt.plot(x, y)

plt.title("Figure Size Example")

plt.show()
```

```python
# ==========================================
# Customize Line Style
# ==========================================
x = [1,2,3,4,5]
y = [3,5,7,9,11]

plt.plot(
    x,
    y,
    color="red",
    linestyle="--",
    linewidth=3,
    marker="o",
    markersize=8
)

plt.title("Customized Line")

plt.show()
```

```python
# ==========================================
# Set Axis Limits
# ==========================================
x = [1,2,3,4,5]
y = [10,20,15,30,25]

plt.plot(x, y)

plt.xlim(0, 6)
plt.ylim(0, 35)

plt.title("Axis Limits")

plt.show()
```

```python
# ==========================================
# Customize Tick Marks
# ==========================================
x = [1,2,3,4,5]
y = [5,10,15,20,25]

plt.plot(x, y)

plt.xticks([1,2,3,4,5])
plt.yticks([0,5,10,15,20,25,30])

plt.title("Custom Tick Marks")

plt.show()
```

```python
# ==========================================
# Add Grid and Legend
# ==========================================
months = ["Jan","Feb","Mar","Apr","May"]

sales = [100,120,150,170,200]
profit = [20,25,30,35,40]

plt.plot(months, sales, label="Sales")
plt.plot(months, profit, label="Profit")

plt.grid(True)
plt.legend()

plt.title("Sales and Profit")

plt.show()
```

```python
# ==========================================
# Add Text Annotation
# ==========================================

x = [1,2,3,4,5]
y = [10,18,15,25,30]

plt.plot(x, y, marker="o")

plt.annotate(
    "Highest Value",
    xy=(5,30),
    xytext=(4,33),
    arrowprops=dict(facecolor="black")
)

plt.title("Annotation Example")

plt.show()
```

```python
# ==========================================
# Save Figure
# ==========================================

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x, y)

plt.title("Saved Figure")

plt.savefig("line_plot.png")

plt.show()
```

```python
# ==========================================
# Professional Example
# ==========================================

months = ["Jan","Feb","Mar","Apr","May","Jun"]
sales = [1200,1500,1700,1600,2000,2200]

plt.figure(figsize=(10,5))

plt.plot(
    months,
    sales,
    marker="o",
    linewidth=3,
    color="steelblue",
    label="Sales"
)

plt.title("Monthly Sales Report", fontsize=16)
plt.xlabel("Months")
plt.ylabel("Sales ($)")

plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()

plt.tight_layout()

plt.show()
```
