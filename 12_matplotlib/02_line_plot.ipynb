```python
# ==========================================
# Matplotlib - Line Plot
# ==========================================

import matplotlib.pyplot as plt
```

```python
# ==========================================
# Example 1 – Basic Line Plot
# ==========================================

# Sample Data
x = [1, 2, 3, 4, 5]
y = [12, 18, 15, 25, 30]

plt.plot(x, y)

plt.title("Basic Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```

```python
# ==========================================
# Example 2 – Line Plot with Marker
# ==========================================

x = [1, 2, 3, 4, 5]
y = [5, 10, 8, 15, 20]

plt.plot(
    x,
    y,
    marker='o'
)

plt.title("Line Plot with Markers")

plt.show()
```

```python
# ==========================================
# Example 3 – Change Line Color and Style
# ==========================================

x = [1, 2, 3, 4, 5]
y = [3, 6, 9, 12, 15]

plt.plot(
    x,
    y,
    color="red",
    linestyle="--",
    linewidth=3
)

plt.title("Customized Line")

plt.show()
```

```python
# ==========================================
# Example 4 – Different Marker Styles
# ==========================================

x = [1,2,3,4,5]
y = [4,8,6,12,10]

plt.plot(
    x,
    y,
    marker="*",
    markersize=12,
    color="green"
)

plt.title("Star Marker")

plt.show()
```

```python
# ==========================================
# Example 5 – Multiple Line Plots
# ==========================================

x = [1,2,3,4,5]

python = [70,75,80,85,90]
java = [65,70,72,78,82]

plt.plot(x, python, label="Python")
plt.plot(x, java, label="Java")

plt.title("Comparison of Scores")
plt.xlabel("Test Number")
plt.ylabel("Score")

plt.legend()

plt.show()
```

```python
# ==========================================
# Example 6 – Adding Grid
# ==========================================

x = [1,2,3,4,5]
y = [2,5,8,11,14]

plt.plot(x, y)

plt.grid(True)

plt.title("Line Plot with Grid")

plt.show()
```

```python
# ==========================================
# Example 7 – Different Line Styles
# ==========================================

x = [1,2,3,4,5]
y = [1,4,2,5,3]

plt.plot(x, y, linestyle="-", label="Solid")
plt.plot(x, [2,3,5,7,9], linestyle="--", label="Dashed")
plt.plot(x, [3,5,4,6,8], linestyle=":", label="Dotted")
plt.plot(x, [4,6,7,8,10], linestyle="-.", label="Dash Dot")

plt.legend()

plt.title("Different Line Styles")

plt.show()
```

```python
# ==========================================
# Example 8 – Complete Professional Example
# ==========================================

import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [1200, 1500, 1700, 1600, 2000, 2300]

plt.figure(figsize=(8,5))

plt.plot(
    months,
    sales,
    color="blue",
    linewidth=3,
    marker="o",
    markersize=8,
    label="Sales"
)

plt.title("Monthly Sales Report")
plt.xlabel("Months")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.legend()

plt.show()
```
