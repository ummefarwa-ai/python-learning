```python
# ==========================================
# Matplotlib - Pie Chart
# ==========================================

import matplotlib.pyplot as plt
```

```python
# ==========================================
# Basic Pie Chart
# ==========================================
labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels)

plt.title("Programming Language Popularity")

plt.show()
```

```python
# ==========================================
# Display Percentages
# ==========================================
labels = ["A", "B", "C", "D"]
values = [30, 25, 20, 25]

plt.pie(
    values,
    labels=labels,
    autopct="%1.1f%%"
)

plt.title("Percentage Distribution")

plt.show()
```

```python
# ==========================================
# Change Colors
# ==========================================
labels = ["Apple", "Banana", "Orange", "Mango"]
sales = [35, 20, 25, 20]

colors = ["red", "yellow", "orange", "green"]

plt.pie(
    sales,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%"
)

plt.title("Fruit Sales")

plt.show()
```

```python
# ==========================================
# Explode a Slice
# ==========================================
labels = ["Python", "Java", "C++", "JavaScript"]
values = [50, 20, 15, 15]

explode = [0.1, 0, 0, 0]

plt.pie(
    values,
    labels=labels,
    explode=explode,
    autopct="%1.1f%%"
)

plt.title("Highlighting Python")

plt.show()
```

```python
# ==========================================
# Add Shadow
# ==========================================
labels = ["Chrome", "Edge", "Firefox", "Safari"]
users = [65, 15, 10, 10]

plt.pie(
    users,
    labels=labels,
    shadow=True,
    autopct="%1.1f%%"
)

plt.title("Browser Usage")

plt.show()
```

```python
# ==========================================
# Start Angle
# ==========================================
labels = ["North", "South", "East", "West"]
sales = [30, 20, 25, 25]

plt.pie(
    sales,
    labels=labels,
    startangle=90,
    autopct="%1.1f%%"
)

plt.title("Regional Sales")

plt.show()
```

```python
# ==========================================
# Donut Chart
# ==========================================
labels = ["Rent", "Food", "Transport", "Savings"]
expenses = [40, 25, 15, 20]

plt.pie(
    expenses,
    labels=labels,
    autopct="%1.1f%%"
)

centre_circle = plt.Circle((0, 0), 0.70, fc="white")

fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Monthly Expenses (Donut Chart)")

plt.show()
```

```python
# ==========================================
# Professional Example
# ==========================================
departments = ["AI", "Software", "Cyber Security", "Data Science"]
students = [120, 90, 60, 80]

colors = ["gold", "skyblue", "lightgreen", "orange"]

plt.figure(figsize=(7,7))

plt.pie(
    students,
    labels=departments,
    colors=colors,
    autopct="%1.1f%%",
    startangle=140,
    shadow=True
)

plt.title("Student Distribution by Department")

plt.show()
```
