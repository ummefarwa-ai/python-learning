```python
# ==========================================
# Matplotlib - Histogram
# ==========================================

import matplotlib.pyplot as plt
import numpy as np
```

```python
# ==========================================
# Basic Histogram
# ==========================================

data = [45, 50, 52, 48, 49, 51, 53, 47, 46, 50]

plt.hist(data)

plt.title("Basic Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
# ==========================================
# Histogram with Bins
# ==========================================

data = [45, 50, 52, 48, 49, 51, 53, 47, 46, 50]

plt.hist(data, bins=5)

plt.title("Histogram with 5 Bins")

plt.show()
```

```python
# ==========================================
# Change Histogram Color
# ==========================================

data = np.random.randint(1, 100, 100)

plt.hist(
    data,
    bins=10,
    color="skyblue"
)

plt.title("Histogram with Color")

plt.show()
```

```python
# ==========================================
# Add Edge Color
# ==========================================

data = np.random.normal(50, 10, 200)

plt.hist(
    data,
    bins=15,
    color="green",
    edgecolor="black"
)

plt.title("Histogram with Edge Color")

plt.show()
```

```python
# ==========================================
# Relative Frequency (Density)
# ==========================================

data = np.random.normal(60, 15, 500)

plt.hist(
    data,
    bins=20,
    density=True,
    color="orange"
)

plt.title("Density Histogram")

plt.show()
```

```python
# ==========================================
# Different Bin Sizes
# ==========================================

data = np.random.normal(100, 20, 300)

plt.figure(figsize=(8,5))

plt.hist(
    data,
    bins=30,
    color="purple",
    edgecolor="white"
)

plt.title("Histogram with More Bins")

plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
# ==========================================
# Transparent Histogram
# ==========================================

data = np.random.normal(50, 12, 200)

plt.hist(
    data,
    bins=15,
    alpha=0.6,
    color="red"
)

plt.title("Transparent Histogram")

plt.show()
```

```python
# ==========================================
# Compare Two Histograms
# ==========================================
group1 = np.random.normal(60, 8, 300)
group2 = np.random.normal(75, 10, 300)

plt.hist(group1, bins=20, alpha=0.6, label="Group 1")
plt.hist(group2, bins=20, alpha=0.6, label="Group 2")

plt.title("Comparison of Two Groups")

plt.legend()

plt.show()
```

```python
# ==========================================
# Professional Example
# ==========================================

exam_scores = np.random.normal(70, 10, 500)

plt.figure(figsize=(8,5))

plt.hist(
    exam_scores,
    bins=20,
    color="steelblue",
    edgecolor="black"
)

plt.title("Distribution of Exam Scores")
plt.xlabel("Marks")
plt.ylabel("Number of Students")

plt.grid(axis="y", linestyle="--")

plt.show()
```
