```python
# ==========================================
# Matplotlib - Bar Charts
# ==========================================

import matplotlib.pyplot as plt
```

```python
# ==========================================
# Basic Bar Chart
# ==========================================
students = ["Ali", "Ahmed", "Sara", "Fatima"]
marks = [80, 75, 90, 85]

plt.bar(students, marks)

plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

```python
# ==========================================
# Change Bar Color
# ==========================================
subjects = ["Math", "Physics", "English", "CS"]
marks = [88, 76, 91, 95]

plt.bar(subjects, marks, color="green")

plt.title("Subject Marks")

plt.show()
```

```python
# ==========================================
# Horizontal Bar Chart
# ==========================================
languages = ["Python", "Java", "C++", "JavaScript"]
users = [95, 80, 70, 85]

plt.barh(languages, users, color="orange")

plt.title("Programming Language Popularity")

plt.show()
```

```python
# ==========================================
# Different Colors for Each Bar
# ==========================================

fruits = ["Apple", "Banana", "Orange", "Mango"]
sales = [45, 70, 50, 90]

colors = ["red", "yellow", "orange", "green"]

plt.bar(fruits, sales, color=colors)

plt.title("Fruit Sales")

plt.show()
```

```python
# ==========================================
# Change Bar Width
# ==========================================

months = ["Jan", "Feb", "Mar", "Apr"]
sales = [120, 180, 160, 200]

plt.bar(months, sales, width=0.4)

plt.title("Monthly Sales")

plt.show()
```

```python
# ==========================================
# Add Grid
# ==========================================

cities = ["Lahore", "Karachi", "Islamabad", "Multan"]
population = [11, 16, 2, 2.1]

plt.bar(cities, population)

plt.grid(axis="y")

plt.title("Population (Millions)")

plt.show()
```

```python
# ==========================================
# Grouped Bar Chart
# ==========================================

import numpy as np

subjects = ["Math", "Physics", "English"]

boys = [80, 75, 90]
girls = [85, 82, 95]

x = np.arange(len(subjects))
width = 0.35

plt.bar(x - width/2, boys, width, label="Boys")
plt.bar(x + width/2, girls, width, label="Girls")

plt.xticks(x, subjects)

plt.title("Marks Comparison")
plt.xlabel("Subjects")
plt.ylabel("Marks")

plt.legend()

plt.show()
```

```python
# ==========================================
# Display Values on Bars
# ==========================================

students = ["Ali", "Ahmed", "Sara", "Fatima"]
marks = [80, 75, 90, 85]

bars = plt.bar(students, marks)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 1,
        height,
        ha="center"
    )

plt.title("Student Marks")

plt.show()
```

```python
# ==========================================
# Professional Example
# ==========================================

courses = ["Python", "NumPy", "Pandas", "Matplotlib"]
hours = [20, 15, 18, 12]

plt.figure(figsize=(8,5))

plt.bar(
    courses,
    hours,
    color="steelblue"
)

plt.title("Study Hours Per Topic")
plt.xlabel("Topics")
plt.ylabel("Hours")

plt.grid(axis="y", linestyle="--")

plt.show()
```
