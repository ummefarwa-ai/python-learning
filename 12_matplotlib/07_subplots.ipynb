```python
# ==========================================
# Matplotlib - Subplots
# ==========================================

import matplotlib.pyplot as plt
import numpy as np
```

```python
# ==========================================
# Two Plots in One Figure (1 Row, 2 Columns)
# ==========================================
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.figure(figsize=(10,4))

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("First Plot")

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Second Plot")

plt.tight_layout()

plt.show()
```

```python
# ==========================================
# Four Subplots (2 × 2)
# ==========================================
x = np.arange(1, 6)

plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.plot(x, x)
plt.title("Line Plot")

plt.subplot(2,2,2)
plt.bar(x, x)
plt.title("Bar Chart")

plt.subplot(2,2,3)
plt.scatter(x, x*2)
plt.title("Scatter Plot")

plt.subplot(2,2,4)
plt.hist(np.random.randn(100))
plt.title("Histogram")

plt.tight_layout()

plt.show()
```

```python
# ==========================================
# Shared X-axis
# ==========================================

x = np.arange(1, 11)
y1 = x
y2 = x ** 2

fig, ax = plt.subplots(2, 1, sharex=True, figsize=(7,6))

ax[0].plot(x, y1)
ax[0].set_title("Linear")

ax[1].plot(x, y2)
ax[1].set_title("Quadratic")

plt.tight_layout()

plt.show()
```

```python
# ==========================================
# Using fig and ax
# ==========================================
x = [1,2,3,4,5]
y = [10,20,15,25,30]

fig, ax = plt.subplots(figsize=(7,5))

ax.plot(x, y, marker="o")

ax.set_title("Using Figure and Axes")
ax.set_xlabel("X Values")
ax.set_ylabel("Y Values")

plt.show()
```

```python
# ==========================================
# Multiple Line Plots
# ==========================================

months = ["Jan", "Feb", "Mar", "Apr", "May"]

sales = [120,150,170,160,200]
profit = [25,30,35,32,45]

fig, ax = plt.subplots(1,2, figsize=(10,4))

ax[0].plot(months, sales)
ax[0].set_title("Sales")

ax[1].plot(months, profit, color="green")
ax[1].set_title("Profit")

plt.tight_layout()

plt.show()
```

```python
# ==========================================
# Different Chart Types Together
# ==========================================

categories = ["A", "B", "C", "D"]
values = [25, 40, 35, 50]

fig, ax = plt.subplots(2,2, figsize=(9,7))

ax[0,0].bar(categories, values)
ax[0,0].set_title("Bar Chart")

ax[0,1].pie(values, labels=categories, autopct="%1.1f%%")
ax[0,1].set_title("Pie Chart")

ax[1,0].scatter([1,2,3,4], [4,6,5,8])
ax[1,0].set_title("Scatter Plot")

ax[1,1].hist(np.random.normal(50,10,200))
ax[1,1].set_title("Histogram")

plt.tight_layout()

plt.show()
```

```python
# ==========================================
# Professional Dashboard
# ==========================================

np.random.seed(42)

sales = np.random.randint(100, 300, 12)
profit = np.random.randint(20, 80, 12)

months = ["Jan","Feb","Mar","Apr","May","Jun",
          "Jul","Aug","Sep","Oct","Nov","Dec"]

fig, ax = plt.subplots(2,1, figsize=(10,8))

ax[0].plot(months, sales, marker="o")
ax[0].set_title("Monthly Sales")

ax[1].bar(months, profit)
ax[1].set_title("Monthly Profit")

plt.tight_layout()

plt.show()
```
