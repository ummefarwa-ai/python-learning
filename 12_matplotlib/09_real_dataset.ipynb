```python
# ==========================================
# Matplotlib - Real Dataset Visualization
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ==========================================
# Create a Dataset
# ==========================================

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [1200, 1500, 1800, 1700, 2100, 2400],
    "Profit": [300, 350, 420, 400, 500, 600]
}

df = pd.DataFrame(data)

print(df)
```

```python
# ==========================================
#Line Plot
# ==========================================
plt.figure(figsize=(8,5))

plt.plot(df["Month"], df["Sales"], marker="o")

plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.show()
```

```python
# ==========================================
# Bar Chart
# ==========================================
plt.figure(figsize=(8,5))

plt.bar(df["Month"], df["Sales"])

plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
# ==========================================
# Profit Line Plot
# ==========================================
plt.figure(figsize=(8,5))

plt.plot(
    df["Month"],
    df["Profit"],
    marker="s",
    color="green"
)

plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")

plt.grid(True)

plt.show()
```

```python
# ==========================================
# Sales vs Profit
# ==========================================
plt.figure(figsize=(8,5))

plt.plot(
    df["Month"],
    df["Sales"],
    marker="o",
    label="Sales"
)

plt.plot(
    df["Month"],
    df["Profit"],
    marker="s",
    label="Profit"
)

plt.title("Sales vs Profit")
plt.xlabel("Month")
plt.ylabel("Amount")

plt.legend()
plt.grid(True)

plt.show()
```

```python
# ==========================================
# Scatter Plot
# ==========================================
plt.figure(figsize=(8,5))

plt.scatter(
    df["Sales"],
    df["Profit"],
    s=120
)

plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.grid(True)

plt.show()
```

```python
# ==========================================
# Pie Chart
# ==========================================
plt.figure(figsize=(6,6))

plt.pie(
    df["Sales"],
    labels=df["Month"],
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Sales Distribution")

plt.show()
```

```python
# ==========================================
# Professional Dashboard
# ==========================================
fig, ax = plt.subplots(2, 2, figsize=(12,8))

# Line Plot
ax[0,0].plot(df["Month"], df["Sales"], marker="o")
ax[0,0].set_title("Sales Trend")

# Bar Chart
ax[0,1].bar(df["Month"], df["Profit"])
ax[0,1].set_title("Monthly Profit")

# Scatter Plot
ax[1,0].scatter(df["Sales"], df["Profit"])
ax[1,0].set_title("Sales vs Profit")

# Pie Chart
ax[1,1].pie(
    df["Sales"],
    labels=df["Month"],
    autopct="%1.1f%%"
)
ax[1,1].set_title("Sales Distribution")

plt.tight_layout()

plt.show()
```

```python
# ==========================================
# Save the Figure
# ==========================================
plt.figure(figsize=(8,5))

plt.plot(
    df["Month"],
    df["Sales"],
    marker="o"
)

plt.title("Monthly Sales")

plt.savefig("monthly_sales.png")

plt.show()
```
