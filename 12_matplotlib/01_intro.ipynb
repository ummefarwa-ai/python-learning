```python
# ==========================================
# Matplotlib - Introduction
# ==========================================

# Import the required library
import matplotlib.pyplot as plt

# Sample Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Create a simple line plot
plt.plot(x, y)

# Add title
plt.title("My First Matplotlib Graph")

# Label x-axis
plt.xlabel("X Values")

# Label y-axis
plt.ylabel("Y Values")

# Display the graph
plt.show()
```

```python
# ==========================================
# Example 2 – Changing Figure Size
# ==========================================
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Change figure size
plt.figure(figsize=(8,5))

plt.plot(x,y)

plt.title("Figure Size Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
# ==========================================
# Example 3 – Different Line Styles
# ==========================================


import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [5,7,9,11,13]

plt.plot(
    x,
    y,
    color="blue",
    linestyle="--",
    linewidth=2,
    marker="o"
)

plt.title("Customized Line Plot")

plt.show()
```

```python
# ==========================================
# Example 4 – Multiple Lines
# ==========================================

import matplotlib.pyplot as plt

x = [1,2,3,4,5]

y1 = [2,4,6,8,10]
y2 = [1,3,5,7,9]

plt.plot(x,y1,label="Even Pattern")
plt.plot(x,y2,label="Odd Pattern")

plt.title("Multiple Lines")
plt.legend()

plt.show()
```

```python
# ==========================================
# Example 5 – Adding Grid
# ==========================================

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,8,6,4,2]

plt.plot(x,y)

plt.grid(True)

plt.title("Graph with Grid")

plt.show()
```
